# Web Scraping Of Flipkart Website of Phone Using bs4:-

```python
import bs4
import requests
```

```python
url='https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
data=requests.get(url)
data
```

```python
soup=bs4.BeautifulSoup(data.text,'html.parser')
print(soup.prettify())
```

```python
mobiles=soup.find_all('div',{'class':'_4rR01T'})
len(mobiles)
```

```python
name=soup.find_all('div',{'class':'_4rR01T'})
print(name[0].text)
```

```python
price=soup.find_all('div',{'class':'_30jeq3 _1_WHN1'})
print(price[0].text)
len(name)
```

```python
rating=soup.find_all('div',{'class':'_3LWZlK'})
print(rating[0].text)
len(rating)
```

```python
nm=[]
pr=[]
rt=[]
for i in name:
    nm.append(i.text)
    
for i in price:
    pr.append(i.text)
    
for i in range(len(nm)):
    rt.append(rating[i].text)
```

```python
data={'Name':nm,'Rating':rt,'Price':pr}
import pandas as pd
df=pd.DataFrame(data)
df.head()
```

```python
df.to_csv('Mobiles.csv',index=False)
```
